## Remove Multicollinearity
- 다중공선성 : 하나의 독립변수가 다른 여러 개의 독립변수들로 잘 예측되는 경우
- 분산팽창계수(VIF) > 10 변수 제거
- 종속 변수가 0 또는 1이므로 로지스틱 회귀를 사용해야하지만,
- 종속 변수가 명목 변수이므로 linear regression를 사용하여 VIF를 구함

```python
df1 = df.copy(deep=False)

del df1["ID"]


input_col =['sex', 'age','N_WATER', 'N_PROT', 'N_FAT', 'N_SFA', 'N_MUFA', 'N_PUFA', 'N_N3', 'N_N6', 'N_CHOL', 'N_CHO', 
            'N_TDF', 'N_SUGAR', 'N_CA', 'N_PHOS', 'N_FE', 'N_NA', 'N_K', 'N_VA', 'N_VA_RAE', 'N_CAROT', 'N_RETIN', 'N_B1', 'N_B2', 'N_NIAC']

df1.to_csv('DF2_dg_no2.csv', index = None)

df1.shape
```

### Modeling

```python
import statsmodels.api as sm

logit1 = sm.Logit(df1[['DF2_dg']],input_data)
result1 = logit1.fit()
```

```python
odd1 = np.exp(result1.params) < 1.03
odd2 = np.exp(result1.params) > 0.97

print(odd1 & odd2)
```

### Model Analysis

```python
print(result1.summary2())  

print(np.exp(result1.params))
```

- coef(편회귀계수) : 값이 양수이면 컬럼의 값이 커질수록 종속변수가 1일 확률이 높아짐
- coef(편회귀계수) : 값이 음수이면 컬럼의 값이 커질수록 종속변수가 0일 확률이 높아짐
    - coef > 0 : N_PROT, N_TDF
    - coef = 0 : N_SUGAR, N_VA_RAE
    - coef < 0 : N_WATER, N_PUFA, N_CHOL, N_CHO, N_CA, N_FE, N_NA, N_B1, N_B2, N_NIAC
    

- P>|t|(유의확률) : (95%의 신뢰도를 가져야 유의미하다고 판단) 독립변수의 유의확률은 0.05보다 작은 값이여야함.
    - P>|t| < 0.05 : N_WATER, N_CHOL, N_CHO, N_FE, N_NA, N_B1, N_B2
    

- OR가 1을 기준으로 큰지 작은지를 파악하여 종속변수에 미치는 영향의 방향을 파악
- OR이 1에서 멀리 떨어질수록 종속변수와의 관계가 강함
    - N_B1, N_B2
